# Classification of normal and abnormal cells in the blood

```python
import numpy as np 
import pandas as pd
import os
import cv2
from tensorflow.keras.utils import  plot_model
```

```python
path='../input/cancer-data/m1/train'

file=os.listdir('../input/cancer-data/m1/train')
# هنا نبدا في عمليه الفصل للصور 
features=[] #to store images
labels=[] #to store labels
for img in file:

    label=int(img[:3]) #extracting labels from the image
    labels.append(label)
    img_array=cv2.imread(os.path.join(path,img),cv2.IMREAD_COLOR)
    img_array=cv2.resize(img_array,(220,220))
    features.append(img_array)
```

```python
# Here we list the main categories of data
dict_1={'000': 'normal', '001': 'not_normal'} 
```

```python
print(set(labels))
```

```python
list_labels=[]
for i in labels:
    new_label=dict_1['00'+str(i)]
    list_labels.append(new_label)
print(list_labels[:2])   
```

# **Here we are working to review part of the data with the class / category of each image.**

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,20))
import seaborn as sns
fig,ax=plt.subplots(3,2)
fig.set_size_inches(12,12)
for i in range(3):
    for j in range (2):
        l=np.random.randint(0,len(labels))
        #print(l)
        ax[i,j].imshow(features[l])
        ax[i,j].set_title( "cell type: " +str(list_labels[l]))
        #print()
plt.axis('off')        
plt.tight_layout()
```
